```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_theme(style="whitegrid", palette="muted", context="notebook")

plt.rcParams['axes.titlesize'] = 16  # Default font size for plot titles
plt.rcParams['axes.labelsize'] = 14  # Default font size for x and y labels

# Sample data
data = {'Category': ['A', 'B', 'C', 'D', 'E', 'F'],
        'Values': [5, 7, 3, 9, 4, 6]}

df = pd.DataFrame(data)

# Set the plot size
plt.figure(figsize=(15, 6))

# Create a bar plot
sns.barplot(x='Category', y='Values', data=df)  # Change 'viridis' to other palettes

# Add labels and title
plt.title('Sample Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')

# Show the plot
plt.tight_layout()  # Adjust layout to fit everything
plt.show()
```
